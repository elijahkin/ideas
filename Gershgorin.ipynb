```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
```

```python
# Returns the centers and radii of the Gershgorin circles of A
def gershgorin(A):
  centers = np.diagonal(A)
  radii = np.sum(np.absolute(A), axis=1) - np.absolute(centers)

  return zip(np.column_stack([centers.real, centers.imag]), radii)
```

```python
# Returns the eigenvalues of A
def eigenvalues(A):
  return np.linalg.eig(A)[0]
```

```python
def gershgorin_plot(A):
  # Plotting the Gershgorin circles
  circles = [plt.Circle(center, radius) for center, radius in gershgorin(A)]

  fig, ax = plt.subplots()

  coll = PatchCollection(circles, edgecolor='black', facecolor='none')
  ax.add_collection(coll)

  ax.set_aspect('equal')
  ax.margins(0)

  # Plotting the eigenvalues of A
  eigenvals = eigenvalues(A)
  plt.scatter(eigenvals.real, eigenvals.imag, color='red')
  plt.show()
```

```python
A = np.array([[6j, 0, -2],
              [0, -1j, -3],
              [1, 2, -2]])

gershgorin_plot(A)
```

```python
# Generate random complex-valued n-by-n matrix
n = 2
A = np.random.randn(n, n) + 1j * np.random.randn(n, n)

gershgorin_plot(A)
print(A)
```
