```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_2d(func, polar=False, bound=None):
  fig = plt.figure(figsize=(4,4))
  ax = fig.add_subplot()
  if polar:
      r, theta = func
      x = r * np.cos(theta)
      y = r * np.sin(theta)
  else:
    x, y, z = func
  ax.plot(x, y, '-')
  if bound:
    plt.xlim([-bound, bound])
    plt.ylim([-bound, bound])
  else:
    plt.axis('equal')
  plt.grid()
  plt.show()

def plot_2d_new(func, polar=False, bound=None, t_min=0, t_max=2*np.pi, color='-'):
  fig = plt.figure(figsize=(4,4))
  ax = fig.add_subplot()

  t = np.linspace(t_min, t_max, (t_max-t_min)/2*np.pi * 10**5)

  if polar:
      r, theta = func
      x = r * np.cos(theta)
      y = r * np.sin(theta)
  else:
    x, y, z = func
  ax.plot(x, y, '-')
  if bound:
    plt.xlim([-bound, bound])
    plt.ylim([-bound, bound])
  else:
    plt.axis('equal')
  plt.grid()
  plt.show()

def plot_3d(func):
  fig = plt.figure(figsize=(4,4))
  ax = fig.add_subplot(projection='3d')
  x, y, z = func
  ax.plot(x, y, z, '-')
  plt.axis('equal')
  plt.show()  
```

```python
def helix(a, b, num):
  t = np.linspace(0, 8*np.pi, num)

  x = a * np.cos(t)
  y = a * np.sin(t)
  z = b * t
  return x, y, z

plot_3d(helix(5, 1, 10**5))
```

```python
def torus_knot(p, q, num):
  phi = np.linspace(0, 2*np.pi, num)
  r = np.cos(q * phi) + 2

  x = r  * np.cos(p * phi)
  y = r  * np.sin(p * phi)
  z = -1 * np.sin(q * phi)
  return x, y, z

plot_3d(torus_knot(3, -7, 10**5))
```

```python
plot_2d(torus_knot(3, -7, 10**5))
```

```python
def butterfly_curve(num):
  t = np.linspace(0, 12*np.pi, num)
  x = np.sin(t) * (np.exp(np.cos(t)) * 2*np.cos(4*t) - np.sin(t/12)**5)
  y = np.cos(t) * (np.exp(np.cos(t)) * 2*np.cos(4*t) - np.sin(t/12)**5)
  return x, y, 0

plot_2d(butterfly_curve(10**5))
```

```python
def astroid(a, num):
  t = np.linspace(0, 2*np.pi, num)
  x = a * np.cos(t)**3
  y = a * np.sin(t)**3
  return x, y, 0

plot_2d(astroid(1, 10**5))
```

```python
def cochleoid(a, num):
  theta = np.linspace(-8*np.pi, 8*np.pi, num)
  r = a * np.sin(theta) / theta
  return r, theta

plot_2d(cochleoid(1, 10**5), polar=True)
```

```python
# https://www.2dcurves.com/roulette/rouletter.html
def rhodonea(c, num):
  theta = np.linspace(0, 4*np.pi, num)
  r = np.sin(c * theta)
  return r, theta
```

```python
# Trifolium
plot_2d(rhodonea(3, 10**5), polar=True)
```

```python
# Quadrifolium
plot_2d(rhodonea(2, 10**5), polar=True)
```

```python
# Dürer Folium
plot_2d(rhodonea(1/2, 10**5), polar=True)
```

```python
def equiangular_spiral(a, b, num):
  theta = np.linspace(-np.pi, 2*np.pi, num)
  r = a * np.exp(theta / np.tan(b))
  return r, theta

plot_2d(equiangular_spiral(2, 1, 10**5), polar=True)
```

```python
# Lissajous Curve
def lissajous_curves(a, b, c, n, num):
  t = np.linspace(0, 2*np.pi, num)
  x = a * np.sin(n*t+c)
  y = b * np.sin(t)
  return x, y, 0

plot_2d(lissajous_curves(1, 1, 1, 3, 10**5))
```

```python
# Archimedean Spiral
def archimedean_spiral(a, num):
  theta = np.linspace(0, 8*np.pi, num)
  r = a * theta
  return r, theta

plot_2d(archimedean_spiral(1, 10**5), polar=True)
```

```python
# Blancmange Curve
def blancmange(num):
  x = np.linspace(0, 1, num)
  y = 0
  for n in range(64):
    s = np.modf(2**n * x)[0]
    s = np.minimum(s, 1-s)
    y += (s / 2**n)
  return x, y, 0

plot_2d(blancmange(10**5))
```

```python
def epi_spiral(a, num):
  theta = np.linspace(-np.pi-1.5, np.pi+1.5, num)
  r = 1 / np.cos(a * theta)
  return r, theta
```

```python
# Trisectrix of Maclaurin
plot_2d(epi_spiral(1/3, 10**5), polar=True, bound=8)
```
