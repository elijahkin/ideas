```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def plot_mat(A):
  plt.imshow(A)
  plt.tick_params(axis='x', bottom=False, labelbottom=False)
  plt.tick_params(axis='y', left=False, labelleft=False)
  plt.show()
```

```python
t = time.time()

n = 2**5
pascal = np.ones((n, n))

for row in range(1, n):
  for col in range(1, n):
    pascal[row][col] = (pascal[row-1][col] + pascal[row][col-1]) % 2

print(time.time()-t)
plot_mat(pascal)
```

```python
t = time.time()

n = 2**5

pascal = np.zeros((n, n), dtype=np.int32)
pascal[0,:] = 1
pascal[:,0] = 1

for i in range(1, n):
    pascal[i, 1:] = np.logical_xor(pascal[i-1, :-1], pascal[i, :-1])
    pascal[1:, i] = np.logical_xor(pascal[:-1, i-1], pascal[:-1, i])

pascal %= 2

print(time.time()-t)
plot_mat(pascal)
```
