```python
import math, csv

def is_prime(n):
    if n < 0:
        return False
    if n == 2 or n == 5:
        return True
    if (n % 10) in [0, 2, 4, 5, 6, 8] or n == 1:
        return False
    for i in range(3, int(n**0.5+1)):
        if n % i == 0:
            return False
    return True
```

```python
# Problem 3
# 6857

def largest_prime_factor(n):
    d = 2
    while d != n:
        if (n % d == 0):
            # Recurse on n without smallest prime factor
            return largest_prime_factor(n // d)
        else:
            d += 1
    # If d == n, only the largest prime factor is left
    return n

print(largest_prime_factor(600851475143))
```

```python
# Problem 10
# 142913828922

# Decent algorithm for finding all primes below n
# https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes
def sieve_of_eratosthenes(n):
    prime = dict()
    for i in range(2, n+1):
      prime[i] = True

    for i in prime:
        factors = range(i, n+1, i)
        for f in factors[1:]:
            prime[f] = False

    return [i for i in prime if prime[i]==True]

print(sum(sieve_of_eratosthenes(2 * 10**6)))
```

```python
# Problem 15
# 137846528820

def lattice_paths(n):
    # Set up matrix
    paths = [[0 for c in range(n+1)] for r in range(n+1)]
    # Iterate through and fill in values
    for x in range(0, n+1):
        for y in range(0, n+1):
            v = 0
            if x == 0 and y == 0:
                v = 1
            if x != 0:
                v += paths[x-1][y]
            if y != 0:
                v += paths[x][y-1]
            paths[x][y] = v
    return paths[n][n]

print(lattice_paths(20))
```

```python
# Problem 69
# 510510

maxN = 10**6
recordRatio = 0
record = 0

def divisors(n):
    divisors = [n]
    for i in range(2, int(math.sqrt(n))+1):
        if n % i == 0:
            divisors += [i, int(n/i)]
    return set(divisors)

def totient(n):
    nDivisors = divisors(n)
    coprimesCount = n
    for d in nDivisors:
        if is_prime(d):
            coprimesCount *= ((d-1)/d)
    return coprimesCount

for n in range(2, maxN):
    ratio = n / totient(n)
    if ratio > recordRatio:
        record = n
        recordRatio = ratio
print(record)

# In retrospect this is very obvious -- The answer will be the greatest 
# primorial under 10**6, that being 510510 = 2*3*5*7*11*13*17*19
# https://en.wikipedia.org/wiki/Primorial
```
